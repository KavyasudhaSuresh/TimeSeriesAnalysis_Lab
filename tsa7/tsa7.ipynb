```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load dataset
file_path = '/content/AAPL.csv'  # Replace with your dataset path
data = pd.read_csv(file_path)

# Convert 'Date' to datetime and set as index
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Use 'Close' prices
ts = data['Close']

# Decompose the time series
# period=30 assumes monthly seasonality for daily data (~1 month)
result = seasonal_decompose(ts, model='additive', period=30)

# Plot the decomposed components
plt.figure(figsize=(14, 10))

plt.subplot(411)
plt.plot(ts, label='Original', color='blue')
plt.legend(loc='upper left')

plt.subplot(412)
plt.plot(result.trend, label='Trend', color='orange')
plt.legend(loc='upper left')

plt.subplot(413)
plt.plot(result.seasonal, label='Seasonality', color='green')
plt.legend(loc='upper left')

plt.subplot(414)
plt.plot(result.resid, label='Residuals', color='red')
plt.legend(loc='upper left')

plt.tight_layout()
plt.suptitle('Time Series Decomposition of AAPL Closing Prices', fontsize=16, y=1.02)
plt.show()
```
